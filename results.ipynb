# Bidirectional Encoder with Fully Connected Layers on the Sentiment Analysis

The following notebook will visualize the model scores received by BOHB or Optuna optimised searched on n_epochs = 7, 12 or 20 accordingly

```python
import pandas as pd
import gensim.downloader as api

import matplotlib.pyplot as plt
import seaborn as sns

from corpusholder import Lang, CorpusHolder
from model import Model, Learner

gensim_model = api.load("glove-wiki-gigaword-100")
print('gensim_model loaded')

def display_result(filename):
    print(filename)
    best_learner = Learner.load(filename, gensim_model)
    plt.plot(best_learner.history['train_loss'], label='train_loss')
    plt.plot(best_learner.history['val_loss'], label='val_loss')
    plt.plot(best_learner.history['val_acc'], label='val_acc')
    plt.title('test_acc=%.4f' % best_learner.history['test_acc'])
    plt.show()
    print(best_learner.config)
    print(best_learner.model.config)
```

### BOHB, n_epochs = 7

```python
display_result('bohb_7.ckpt')
```

### BOHB, n_epochs = 12

```python
display_result('bohb_12.ckpt')
```

### BOHB, n_epochs = 20

```python
display_result('bohb_20.ckpt')
```

### Optuna, n_epochs = 7

```python
display_result('optuna_7.ckpt')
```

### Optuna, n_epochs = 12

```python
display_result('optuna_12.ckpt')
```
